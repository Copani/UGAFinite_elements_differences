# Exercise 1 - Discrete first and second order derivtives

```python
import numpy as np
from numpy import sin, cos, pi
import matplotlib.pyplot as plt
```

```python
T = 0.2
dt = 1e-3

A = 10
nu = 10 #Hz
w = 2 * pi * nu
dt = 1e-3
T = T
t = np.arange(0,T,dt)

def F(t):
    return A * cos(w * t)

def f(t):
    return - A * w * sin(w * t)

def f_for(t):
    return (F(t + dt) - F(t))/dt

def f_back(t):
    return (F(t) - F(t - dt))/dt

def f_both(t):
    return (F(t + dt) - F(t - dt)) / (2 * dt)

def f_dot(t):
    return - A * w**2 * cos(w * t)

def f_dot_discrete(t):
    return (F(t + dt) - 2 * F(t) + F(t - dt)) / (dt ** 2)


plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.plot(t,f(t), linewidth=3,label='$\dot{f}$ closed_form')
plt.plot(t,f_for(t),label='$\dot{f}$ forward')
plt.plot(t,f_back(t),label='$\dot{f}$ backward')
plt.plot(t,f_both(t),label='$\dot{f}$ both')
plt.title('first derivartive, $\Delta t$ = {}'.format(dt))
plt.legend()

plt.subplot(2,2,2)
dt = 1e-2
plt.plot(t,f(t), linewidth=3,label='$\dot{f}$ closed_form')
plt.plot(t,f_for(t),label='$\dot{f}$ forward')
plt.plot(t,f_back(t),label='$\dot{f}$ backward')
plt.plot(t,f_both(t),label='$\dot{f}$ both')
plt.title('first derivative, $\Delta t$ = {}'.format(dt))
plt.legend()


plt.subplot(2,2,3)
dt = 1e-3
plt.plot(t,f_dot(t),label='$\ddot{f}}$ closed_form')
plt.plot(t,f_dot_discrete(t),label='$\ddot{f}$ discrete approx' )
plt.title('second derivative, $\Delta t$ = {}'.format(dt))
plt.legend()

plt.subplot(2,2,4)
dt = 1e-2
plt.plot(t,f_dot(t),label='$\ddot{f}}$ closed_form')
plt.plot(t,f_dot_discrete(t),label='$\ddot{f}$ discrete approx' )
plt.title('second derivative, $\Delta t$ = {}'.format(dt))
plt.legend()


```

# Exercise 2 - The Euler Scheme


```python
c = 2
A = 50
T0 = A
tmax = 30
dt = 0.1
t = np.arange(0,tmax,dt)

def T(t):
    return A * np.e ** -(t/c)

def T_euler(t):
    T = np.zeros(t.shape)
    T[0] = T0
    for i in range(len(t)-1):
        T[i+1] = T[i] - dt * T[i]/c
    return T

plt.plot(t,T(t),label='T closed')
dt = 0.1
t = np.arange(0,tmax,dt)
plt.plot(t,T_euler(t),label='T euler, dt = {}'.format(dt))
dt = 0.5
t = np.arange(0,tmax,dt)
plt.plot(t,T_euler(t),label='T euler, dt = {}'.format(dt))
dt = 3
t = np.arange(0,tmax,dt)
plt.plot(t,T_euler(t),label='T euler, dt = {}'.format(dt))
plt.legend()
dt = 4.1
t = np.arange(0,tmax,dt)
plt.plot(t,T_euler(t),label='T euler, dt = {}'.format(dt))
plt.legend()
```
